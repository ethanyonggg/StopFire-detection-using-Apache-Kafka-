## Producer 2
- Writing a python program that loads all the data from hotspot_AQUA_streaming.csv and randomly (with replacement) feed the data to the stream every n seconds, where n is some random 
  number AQUA is the satellite from NASA that reports latitude, longitude, confidence and surface temperature of a location.

```python
# import statements
from time import sleep
from json import dumps
from kafka3 import KafkaProducer
import random
import pandas as pd
import datetime as dt
import json
```

```python
def publish_message(producer_instance, topic_name, key, value):
    try:
        key_bytes = bytes(key, encoding='utf-8')
        value_str = str(json.dumps(value, sort_keys=True, indent=4))
        value_bytes = bytes(value_str, encoding='utf-8')
        producer_instance.send(topic_name, key = key_bytes, value=value_bytes)
        producer_instance.flush()
        print('Message published successfully. Data: ' + str(value))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=[f'{"10.192.96.162"}:9092'], api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
if __name__ == '__main__':
   
    topic = 'partB2'

    print('Publishing records..')
    producer = connect_kafka_producer()
    latest_date = dt.datetime.now() 
    
    climate = pd.read_csv("hotspot_AQUA_streaming.csv")
    array = []        
    for index, row in climate.iterrows():
        message = row.to_dict()
        message['latitude'] = round(float(message['latitude']), 2)
        message['longitude'] = round(float(message['longitude']), 2)
        message['confidence'] = float(message['confidence'])
        message['surface_temperature_celcius'] = int(message['surface_temperature_celcius'])
        message['producer'] = "producer2"
        array.append(message)
    
    for _ in range(len(array)):
        random_number = random.randrange(0, len(array))
        random_data = array[random_number]
        latest_date += dt.timedelta(seconds=2)  
        # Set date to string 
        random_data['created_date'] = latest_date.strftime('%Y-%m-%d %H:%M:%S')   
        #key = producer2
        publish_message(producer, topic, "producer2", random_data)
        sleep(2)
```
